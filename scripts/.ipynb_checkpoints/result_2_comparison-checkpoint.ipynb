```python
organ = 'Task02_Heart'
organ = 'Task05_Prostate'
all_baselineA0T0 = []
all_baselineA1T0 = []
all_baselineA1T1 = []
all_windowA0T0 = []
all_windowA1T0 = []
all_windowA1T1 = []
all_window_allA0T0 = []
all_window_allA1T0 = []
all_window_allA1T1 = []


file_name = os.listdir(os.path.join("../results/expe_20240510/", organ, 'baselineA0T0'))
file_name = [file for file in file_name if not file.endswith('.png')]
file_name = [file for file in file_name if not file.endswith('.ipynb_checkpoints')]
file_name = [file for file in file_name if not file.endswith('.pth')]
file_name = [file for file in file_name if not file.endswith('_time')]

print(file_name)
print(len(file_name))

for f in file_name:
    baselineA0T0= np.loadtxt('../results/expe_20240510/'+ organ + '/baselineA0T0/' +f, delimiter=',')
    baselineA1T0= np.loadtxt('../results/expe_20240510/'+ organ + '/baselineA1T0/' +f, delimiter=',')
    baselineA1T1= np.loadtxt('../results/expe_20240510/'+ organ + '/baselineA1T1/' +f, delimiter=',')
    windowA0T0= np.loadtxt('../results/expe_20240510/'+ organ + '/windowA0T0/' +f, delimiter=',')
    windowA1T0= np.loadtxt('../results/expe_20240510/'+ organ + '/windowA1T0/' +f, delimiter=',')
    windowA1T1= np.loadtxt('../results/expe_20240510/'+ organ + '/windowA1T1/' +f, delimiter=',')
    window_allA0T0= np.loadtxt('../results/expe_20240510/'+ organ + '/window_allA0T0/' +f, delimiter=',')
    window_allA1T0= np.loadtxt('../results/expe_20240510/'+ organ + '/window_allA1T0/' +f, delimiter=',')
    window_allA1T1= np.loadtxt('../results/expe_20240510/'+ organ + '/window_allA1T1/' +f, delimiter=',')
    
    all_baselineA0T0.append(np.nanmean(baselineA0T0))
    all_baselineA1T0.append(np.nanmean(baselineA1T0))
    all_baselineA1T1.append(np.nanmean(baselineA1T1))
    all_windowA0T0.append(np.nanmean(windowA0T0))
    all_windowA1T0.append(np.nanmean(windowA1T0))
    all_windowA1T1.append(np.nanmean(windowA1T1))
    all_window_allA0T0.append(np.nanmean(window_allA0T0))
    all_window_allA1T0.append(np.nanmean(window_allA1T0))
    all_window_allA1T1.append(np.nanmean(window_allA1T1))
    
    # print(np.nanmean(baselineA0T0), np.nanmean(baselineA1T0), np.nanmean(baselineA1T1), np.nanmean(windowA0T0), np.nanmean(windowA1T0), np.nanmean(windowA1T1),np.nanmean(window_allA0T0),np.nanmean(window_allA1T0),np.nanmean(window_allA1T1))
    
```

```python
from scipy.stats import ttest_rel
t_stat1, t_val1 = ttest_rel(all_baselineA1T1, all_window_allA1T1, alternative = 'less')
t_stat2, t_val2 = ttest_rel(all_baselineA1T1, all_window_allA1T0, alternative = 'less')

print(t_stat1, t_val1)
print(t_stat2, t_val2)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# 仮のデータです。実際のdiceスコアのリストに置き換えてください。
dice_scores = [
    all_baselineA0T0, all_baselineA1T0, all_baselineA1T1,
    all_windowA0T0, all_windowA1T0, all_windowA1T1,
    all_window_allA0T0, all_window_allA1T0, all_window_allA1T1
]

# 群ごとに色を設定
palette = ["red"] * 4 + ["green"] * 4 + ["blue"] * 3

# モデル名
labels = [
    'baselineA0T0', 'baselineA1T0', 'baselineA1T1',
    'windowA0T0', 'windowA1T0', 'windowA1T1',
    'window_allA0T0', 'window_allA1T0', 'window_allA1T1'
]

# ダミーデータを追加して群間にスペースを作る
expanded_scores = []
expanded_labels = []
for i, score in enumerate(dice_scores):
    expanded_scores.append(score)
    expanded_labels.append(labels[i])
    if (i + 1) % 3 == 0 and i < len(dice_scores) - 1:
        # 各群の最後でダミーデータを挿入
        expanded_scores.append([np.nan])
        expanded_labels.append('')

# 箱ひげ図の描画
plt.figure(figsize=(10, 6))
sns.boxplot(data=expanded_scores, palette=palette)
plt.title('Dice Scores by Model')
plt.xlabel('Model')
plt.ylabel('Dice Score')
plt.xticks(range(len(expanded_labels)), expanded_labels, rotation=45)  # x軸ラベルを45度回転
plt.grid(True)
plt.show()
```
